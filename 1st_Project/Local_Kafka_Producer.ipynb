```python
from kafka import KafkaProducer 
from json import dumps 
import time 

# kafka에 실시간으로 데이터를 넣어줄 producer 생성
producer = KafkaProducer(acks=0, compression_type='gzip', 
                         bootstrap_servers=['172.17.0.6:9092','172.17.0.7:9092','172.17.0.8:9092'], # broker ip 입력
                         value_serializer=lambda x: dumps(x).encode('utf-8')
                        )
start = time.time()

# json data를 1 item씩 전송
for data in json_data:
    producer.send('hdtest', data)
    producer.flush() 

print("elapsed :", time.time() - start)

```
